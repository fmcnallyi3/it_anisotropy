Functions
------------------------
 Functions allow for the consolidation of several pieces of code into a single reusable object. The basic syntax for defining a function is shown below.

```python
def function(parameters):
    #code code code
    return ouput
```

The following are examples of some basic functions:

```python
def add(value1, value2):
    total = value1 + value2
    return total
print(add(4,5))
print(add('Python',' rules'))
```

```python
def change_zeroth_item_to_3(parameter):
    parameter[0] = 3
    return parameter
change_zeroth_item_to_3([1,2,3,4,5])
```

Functions are very useful for compartmentalizing our code. Why write dozens or hundreds of lines of interconnecting code when it will end up messy and confused? Many times you will have a complex idea for a solution nd that solution will have a lot of moving parts. Instead of writing one very long piece of code, write simple blocks each with a specific purpose or function. These functions help simplify the process and make debugging easier.

# Flow Control

## If
It is very common that we will want to run specific code for different conditions. In these cases, we use If/Else/Elif statements to run code given for a given condition. The following are examples of each.

```python
!# usr/bin/env python3

a = 10
if a < 20:
    print('a is less than 20')
```

We can even use an If statement to define our main function.

```python
!# usr/bin/env python 3

if __name__ == "__main__":
    #code code code
```

This way to write the main function is preferred when the code we are writing is expected to be exported into another script. Sometimes we run into issues when the main function is executed improperly by another code, so we want to make sure it does not execute. As we can see, this way of writing the main function requires that we have code to execute while the other way did not have this requirement. You will typically need the latter form when writing long codes with multiple parts.


Sometimes we want to specify that two conditions are met before we run our code. We can use the AND conditional to make sure both conditions are met.

## AND/OR

```python
!# usr/bin/env python3

valo = 2
valt = 5
if valo < 5 and valt < 5:
    print('both numbers are less than 5')
```

We can use the OR conditional to run the code when at least one of our conditions are met.

```python
!# usr/bin/env python3

valo = 2
valt = 6
if valo < 5 or valt < 5:
    print('at least one number is less than 5')
```

As we can see, the OR conditional lets our code run, but is not very description on which condition is met. 

While the If statement is very helpful, we will more often than not have multiple conditions we want to run our code on. The following is an example of an If-Else statement.

## If-Else

```python
!# usr/bin/env python 3

a = 15
if a < 20:
    print('a is less than 20')
else:
    print('a is greater than 20')
```

The If-Else statement is great when we have two conditions we want to meet, such as Boolean conditions which are the simpler case for our If-Else conditions. Of course, there may be more than two conditions we need to meet and there are a couple solutions for these, the Elif statement and the Match Case (may be called Switch Case in other languages). Below is an example of the Elif statement.

## Elif

```python
!# usr/bin/env python3

b = 'blue'
if b == 'red':
    print('b is red')
elif b == 'green':
    print('b is green')
elif b == 'blue':
    print('b is blue')
else:
    print('b is not a color')
```

The Elif statement is very versatile for sorting through multiple conditions. When writing an Elif statement it is good practice to finish with an Else statement to avoid bugs. The Else statement at the end of an Elif can be very helpful and descriptive in figuring out the bugs in the code.

A more complex but just as helpful flow controller is the Match Case. The match case is different from the Elif because it checks an input while the If/Else/Elif statements passively check a condition. The following is an example of a Match Case.

## Match Case

```python
!# usr/bin/env python3

def lever(condition):
    match condition:
        case 'condition1':
            print('meets condition 1')
        case 'condition2':
            print('meets condition 2')
        case '_':
            print('invalid condition')
```

As we see above, our Match Case is a function that can take any case and gives an output. Here, we can specify the particular input which is a deeper level of micro managing for our code. While the example has print statements, our cases can execute other functions when a condition is met. 

Now, we do not want to manually execute our code for each data point. The solution to this problem are loops.

# Loops

Instead of running our code over every single single file and every single line, we want to run it once and let it run over one file and let the code take it from there. We do this with loops, more particularly, For and While loops. For loops are analogous to series in calculus. The following is an example of a For loop.

## For Loop

```python
!# usr/env/bin python

i = 0
for i in range(3) :
    i + 1
print(i)
```

We use for loops when we have a known limit we want to run over. Above we use the range() function to run our loop until i = 2. Here, i is our index. A common mistake with For loops is that someone will forget to increase the index in the loop. When the loop reiterates, it checks that the index meets the condition. Another useful conditional for our loop is the "Line in file" conditional which will look over every line in a file and stop when there are no more lines to read. This conditional is certainly helpful when we need to generally read files of different sizes. The following is an example.

```python
!# usr/env/bin python

# file = [path to file]

# for Line in file:
    #code code code
```

Here we specify what file we are looking at. In the next part, you will see a much easier way to incorporate the command line with customized arguments to specify which file to run over. Another loop we use is the While loop, which runs until a more direct boolean condition is reached.

## While Loop

For loops are good when we have a defined limit, such as a specific length. On the other hand, While loops let us run until a boolean condition is met. The following is an example.

```python
!# usr/bin/env python

i = False
while i != True:
    i = True
print("i is True")
```

Like in the For loop, the loop checks the conditional. Where the For loop runs for a defined length, the While loop runs until the conditional is met. For loops are preferred for a few reasons. For loops take less time to run, can run by checking the conditional ahead of the loop, and some personal preferences. The previous look into loops works well in a one-dimensional sense, but we can Nest them to go even further beyond.

## Nesting

Many times we wil be working with multi-dimensional arrays. In order to look through or populate these arrays, we need to loop through each dimension. We do this by nesting our loops, that is, we put a loop inside of a loop. The following is an example.

```python
!# usr/bin/env python

matrix = [[3,3,3],
          [5,5,5],
          [7,7,7]]

for row in matrix:
    for element in row:
        print(element, end =' ')
    print()
```

And there's our matrix. In the next section, we will see a different way of doing this but it is important to understand the basis. Of course, you can also nest If/Else/Elif statements into a loop, like so.

```python
!# usr/bin/env python

for i in range(0,11):
    if i % 2 == 0:
        print(i, 'is even')
```

Here, we simply check if a number is even and reiterate that until we reach double digits. Nesting may get complicated, so when coding or examining code, it is important to step back and look at what each piece does. It is ok to get stuck or confused from time to time. ## Overloading

Overloading is used in object-oriented programming (the language makes and reference 'objects') to write scripts that make and customize objects for us to use. Python, Java, and C++ are examples of object-oriented languages because of their use of functions/methods (which make and reference 'objects'). Overloading occurs when you create two functions with the same name, but different parameters to enter. For example, two functions could be called "add" but the first "add" could have two parameters while the second "add" has three. Unfortunately in Python, overloading is ineffecient since the latest method would be called instead of the one we want (our second "add" would run and probably shout at us).  Let's move on to writing our first Python script and then talk more about things we can use in python. For our first program let's keep it simple and use some ideas we already have worked with. How about we find all the prime numbers up to 100? To do this we need to test if each number between 2 and 100 is divisible by any number smaller than itself. We can do this pretty easily with a list, a few loops, and an if statement.


```python
#!/usr/bin/env python3

numbers0to50=range(2,51)
for num in numbers0to50:
    prime=True
    for chknum in range(2,num):
        if num%chknum==0:
            prime=False
    if prime==True:
        print(num,"is prime")
```

 We can copy this script into our text editor and then save it with a descriptive name in a directory we will use for these script examples. On the other hand, we can save this into a file via jupyter or VSC. Let's name it primenumbers.py, where the .py extension identifies what kind of file it is. Now we have a script and need to run it.

### Running a python script

There are two main ways to run a python script. Firstly, we can give the path to our script as the argument of the python command on the command line in a shell by typing

 python3 primenumbers.py

 Secondly, we can include the so-called shebang line

\#!/usr/bin/env python3

as the first line of our script, which if the file is executable will figure out which python on your system to use and then run the script by typing on the command line

 ./primenumbers.py

 Both methods will produce the same result, so it's really just a matter of preference. Let's run our program by the first method for now by making sure our shell is in the same directory as our script the then running

 python3 primenumbers.py.

 You should see the same result as when running it in the Jupyter notebook.

Coding Style
------------------------------

[[back to top](#top)]

 Before we get too far along for anyone to develop bad habits, let's stop and talk about some style guidelines you should follow as they are laid out in the [Python Enhancement Proposal (PEP)](https://www.python.org/dev/peps/pep-0008/).

### Indentation

Python is different than other languages in that it uses white space to determine what control statement it is inside of. Use 4 spaces for indentation. This is enough space to give your code some visual structure, while leaving room for multiple indentation levels. If you plan on writing scripts in a text editor, most will have settings such that a tab inserts 4 spaces. The compiler will throw errors if it expects code to be indented, make sure to check for proper indentation.

### Line Length

Use up to 79 characters per line of code, and 72 characters for comments. This is a style guideline that some people adhere to and others completely ignore. This used to relate to a limit on the display size of most monitors. Now almost every monitor is capable of showing much more than 80 characters per line. But we often work in terminals, which are not always high-resolution. We also like to have multiple code files open, next to each other. It turns out this is still a useful guideline to follow in most cases. When using functions whose arguments get long, one can simply break after the completion of an argument with a return, python will keep reading in the arguments until it finds a matching ')'. If a string or equation runs too long a backslash ('\') can be used to break it into more then one line.

### Blank Lines

Use single blank lines to break up your code into meaningful blocks. This is helpful for organizing your code and making it easier to read.

### Comments

Use a single space after the pound sign (#) at the beginning of a line. If you are writing more than one paragraph, use an empty line with a pound sign between paragraphs. Make comments concise and easy to read. The point is to communicate to someone who does not know the code what the code does.

### Naming Variables

Name variables and program files using only lowercase letters, underscores, and numbers. Python won't complain or throw errors if you use capitalization, but you will mislead other programmers if you use capital letters in variables at this point. When setting global variables (variables outside functions), it is customary to use capitalization on every character to differential the variable from a local variable (variables inside functions).

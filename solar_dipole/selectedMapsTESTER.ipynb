```python
import healpy as hp
import numpy as np
import matplotlib.pyplot as plt

from glob import glob

from mapFunctions import getMap
```

```python
# FUNCTION
def plotSkyMap(monthIn: int, year: int):
    month = f"{monthIn:02}"
    try:
        solar_map = sorted(glob(f'/data/user/fmcnally/anisotropy/maps/maps_12yr_smt08/IC86-{year}/*solar_????-{month}-??.fits'))
        m = getMap(solar_map, mapName='signal', smooth=5, verbose=False, mask=True)
        hp.mollview(m, min=-7, max=7, title=month)
        plt.show()
    except IndexError:
        print(f'\nFailed to find data for month {month} :(\n')
    except:
        print("Something else failed.")
```

```python
## CALLING THE FUNCTION
plotSkyMap(1, 2011)
```
